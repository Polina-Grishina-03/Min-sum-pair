### Код решения задачи:

```python
def min_sum_pair(arr):
    integers = [x for x in arr if isinstance(x, int)]
    if len(integers) < 2:
        raise ValueError("В массиве должно быть не менее двух целых чисел")
    integers.sort()
    min_pair = (integers[0], integers[1])
    min_sum = integers[0] + integers[1]
    
    return min_pair, min_sum
```

### Тесты для проверки кода:

1. Базовый случай (стандартный ввод)

```python
print(min_sum_pair([3, 1, 4, -2, 0, 5]))
```

2. Массив только с отрицательными числами

```python
print(min_sum_pair([-10, -5, -1, -20]))
```

3. Массив с дубликатами

```python
print(min_sum_pair([2, 2, 1, 1]))
```

4. Массив с нечисловыми элементами (фильтрация)

```python
print(min_sum_pair([3.14, "hello", 5, None, -3, 0]))
```

5. Граничный случай: ровно 2 целых числа

```python
print(min_sum_pair([100, 200, 3.5, "abc"]))
```

6. Ошибка: нет целых чисел

```python
print(min_sum_pair([1.5, "text", None]))
```

7. Ошибка: ровно 1 целое число

```python
print(min_sum_pair([42, 3.14, "x"]))
```

8. Массив с нулями

```python
print(min_sum_pair([0, 0, -1, 1]))
```

9. Большие числа (проверка на переполнение)

```python
print(min_sum_pair([10**9, -10**9, 5, -5]))
```

10. Уже отсортированный массив

```python
print(min_sum_pair([-5, -2, 0, 3, 7]))
```
